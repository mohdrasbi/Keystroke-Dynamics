```python
import time
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/output_vasu.csv", index_col=0)
```

```python
df = df.sort_values(by=['press_time'])
```

```python
df
```

```python
import pandas as pd
info_dict = dict.fromkeys(["q", "a", "z", "1"], "ll")
info_dict.update(dict.fromkeys(["w", "s", "x", "2"], "lr"))
info_dict.update(dict.fromkeys(["e", "d", "c", "3"], "lm"))
info_dict.update(dict.fromkeys(["r", "t", "f", "g", "v", "b", "4", "5"], 'li'))
info_dict.update(dict.fromkeys(["!", "Q", "A", "Z"], 'll_cap'))
info_dict.update(dict.fromkeys(["@", "W", "S", "X"], 'lr_cap'))
info_dict.update(dict.fromkeys(["#", "E", "D", "C"], 'lm_cap'))
info_dict.update(dict.fromkeys(["$", "R", "F", "V", "%", "T", "G", "B"], 'li_cap'))
info_dict.update(dict.fromkeys(["y", "u", 'h', 'j', 'n', 'm', '6', '7'], 'ri'))
info_dict.update(dict.fromkeys(['i', 'k', ',', '8'], 'rm'))
info_dict.update(dict.fromkeys(['.', '9', 'o', 'l'], 'rr'))
info_dict.update(dict.fromkeys(['0', 'p', ';', '/', '[', "'", "`", ']'], 'rl'))
info_dict.update(dict.fromkeys(["^", "Y", "H", "N", "&", "U", "J", "M"], 'ri_cap'))
info_dict.update(dict.fromkeys(["*", "I", "K", "<"], 'rm_cap'))
info_dict.update(dict.fromkeys(["(", "O", "L", ">"], 'rr_cap'))
info_dict.update(dict.fromkeys([")", "P", ":", "?", "{", "}", '"'], 'rl_cap'))
```

```python
'q' in info_dict.keys()
```

```python
features = ["ri_left", "ri_right", "ri_same", "ri_ht", "ri_cap_ht", 
    "rm_left", "rm_right", "rm_same", "rm_ht", "rm_cap_ht", "rr_left", "rr_right", "rr_same", "rr_ht", "rr_cap_ht", "rl_left", "rl_right", 
    "rl_same", "rl_ht", "rl_cap_ht", 'li_left', 'li_right', 'li_same', 'li_ht', 'li_cap_ht', 'lm_left', 'lm_right', 'lm_same', 'lm_ht', 
    'lm_cap_ht', 'lr_left', 'lr_right', 'lr_same', 'lr_ht', 'lr_cap_ht', 'll_left', 'll_right', 'll_same', 'll_ht', 'll_cap_ht', 'cpm', 
    'accuracy', 'sb_dd', 'sb_ht', 'key_sb', "user"]
feat_map = {features[i]: i for i in range(len(features))}
```

```python
length = df.shape[0]
start_time = df.iloc[0]['press_time']
bin_size_const = 10
bin_size = bin_size_const
feat_values = [[] for i in range(len(features))]
new_df = pd.DataFrame(columns=features)
index = 0
backspaces = 0

shift = ["Key.shift", "Key.shift_r"]


for i in range(1, length):
    prev = df.iloc[i-1]
    prev_key = prev['key']
    prev_time = prev['press_time']
    
    curr = df.iloc[i]
    curr_key = curr['key']
    curr_time = curr['press_time']
    
    time_diff = curr_time - start_time
    if curr_key.isupper() or prev_key in shift or prev_key == "Key.backspace"\
        or prev_key == 'Key.enter' or curr_key == 'Key.enter':
        continue
            
            
    if curr_key == "Key.backspace":
        backspaces += 1
        continue
        
    if (curr_key in info_dict.keys()) and (prev_key in info_dict.keys()) and (info_dict[curr_key] == info_dict[prev_key]):
        feat_transition = "{}_same".format(info_dict[curr_key])
        feat_ht = "{}_ht".format(info_dict[curr_key])

        feat_values[feat_map[feat_transition]].append(curr_time - prev_time)
        feat_values[feat_map[feat_ht]].append(curr['hold_time'])
        
    elif curr_key == "Key.space":
        feat_values[feat_map["sb_dd"]].append(curr_time - prev_time)
        feat_values[feat_map["sb_ht"]].append(curr['hold_time'])
        
    elif curr_key in shift:
        feat_ht = "{}_ht".format(info_dict[df.iloc[i+1]['key'].upper()])
        feat_values[feat_map[feat_ht]].append(curr['hold_time'])
        
    elif prev_key == "Key.space":
        feat_transition = 'key_sb'
        feat_ht = "{}_ht".format(info_dict[curr_key])
        
        feat_values[feat_map[feat_transition]].append(curr_time - prev_time)
        feat_values[feat_map[feat_ht]].append(curr['hold_time'])
        
    else:               
        temp = ""
        if info_dict[prev_key][0] == "l":
            temp = "left"
        elif info_dict[prev_key][0] == "r":
            temp = "right"

        feat_transition = "{}_{}".format(info_dict[curr_key], temp)
        feat_ht = "{}_ht".format(info_dict[curr_key])
        feat_values[feat_map[feat_transition]].append(curr_time - prev_time)
        feat_values[feat_map[feat_ht]].append(curr_time - prev_time)

        
    if time_diff >= bin_size:
        cpm = i/bin_size*60
        feat_values[feat_map['cpm']].append(cpm)
        feat_values[feat_map['accuracy']].append(100 - (backspaces/i*100))
        
        new_df.loc[index] = list(map(lambda x: np.mean(x) if len(x) > 0 else np.nan, feat_values))
        index += 1
        bin_size += bin_size_const
        feat_values = [[] for i in range(len(features))]
        backspaces = 0
```
